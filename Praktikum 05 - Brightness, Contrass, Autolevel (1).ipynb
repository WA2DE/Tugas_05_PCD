# Brightness, Contrass, Autolevel Contrass

Import dependency

```python
import numpy as np #  Untuk melakukan operasi matematika pada array atau matriks dalam Python.
import imageio # untuk membaca, menulis, dan memanipulasi file gambar dan video dalam berbagai format
import matplotlib.pyplot as plt # Digunakan untuk membuat plot dan visualisasi data dalam bentuk grafik atau gambar.
```

Membaca gambar

```python
img = imageio.imread("citra.jpg") # mengimport sebuah citra
```

Mendapatkan resolusi dan type dari gambar

```python
img_height = img.shape[0] # menyimpan nilai tinggi citra (jumlah baris piksel) yang dapat diakses menggunakan fungsi shape[0]
img_width = img.shape[1] # menyimpan nilai lebar citra (jumlah kolom piksel) yang dapat diakses menggunakan fungsi shape[1]
img_channel = img.shape[2] # menyimpan jumlah saluran warna citra (misalnya RGB) yang dapat diakses menggunakan fungsi shape[2]
img_type = img.dtype # variabel yang menyimpan tipe data piksel citra
```

# Brightness Grayscale

Membuat variabel img_brightness untuk menampung hasilÂ¶

```python
img_brightness = np.zeros(img.shape, dtype=np.uint8) # Menginisialisasi variabel img_brightness dengan nilai zero array yang memiliki ukuran sama dengan img dan tipe data uint8
```

Melakukan penambahan brightness dengan nilai yg menjadi parameter

```python
def brighter(nilai): # # Merupakan deklarasi fungsi
    for y in range(0, img_height): # Melakukan perulangan sebanyak tinggi gambar
        for x in range(0, img_width): # Melakukan perulangan sebanyak lebar gambar
            red = img[y][x][0] # nilai komponen warna merah
            green = img[y][x][1] # nilai komponen warna hijau
            blue = img[y][x][2] # nilai komponen warna biru
            gray = (int(red) + int(green) + int(blue)) / 3 # menghitung rata-rata grayscale
            gray += nilai # menambahkan nilai kecerahan
            if gray > 255: # # memastikan bahwa nilai grayscale tidak melebihi 255 (nilai maksimum)
                gray = 255
            if gray < 0: # memastikan bahwa nilai grayscale tidak kurang dari 0 (nilai minimum)
                gray = 0
            img_brightness[y][x] = (gray, gray, gray) # menetapkan nilai grayscale baru ke setiap komponen warna (r, g, b) pada setiap piksel gambar
```

Menampilkan beberapa hasil dengan nilai brightness -100 dan 100

```python
brighter(-100) # Memanggil fungsi brighter dengan parameter -100, untuk mengurangi kecerahan gambar
plt.imshow(img_brightness) # Menampilkan gambar hasil pengurangan kecerahan dengan plt.imshow()
plt.title("Brightness -100") # Menampilkan judul gambar hasil pengurangan kecerahan dengan plt.title()
plt.show() # Menampilkan gambar

brighter(100) # Memanggil fungsi brighter dengan parameter 100, untuk menambah kecerahan gambar
plt.imshow(img_brightness) # Menampilkan gambar hasil penambahan kecerahan dengan plt.imshow()
plt.title("Brightness 100") # Menampilkan judul gambar hasil penambahan kecerahan dengan plt.title()
plt.show() # Menampilkan gambar
```

# Brightness RGB

Membuat variabel img_rgbbrightness untuk menampung hasil

```python
img_rgbbrightness = np.zeros(img.shape, dtype=np.uint8) # Mendeklarasikan array img_rgbbrightness yang berukuran sama dengan gambar img, dengan tipe data unsigned integer 8-bit (uint8) yang seluruh elemennya diisi dengan nilai 0
```

Melakukan penambahan brightness dengan nilai yg menjadi parameter

```python
def rgbbrighter(nilai): # Deklarasi fungsi dengan satu parameter nilai
    for y in range(0, img_height): # Melakukan perulangan sebanyak tinggi gambar
        for x in range(0, img_width): # Melakukan perulangan sebanyak lebar gambar
            red = img[y][x][0] # Mendapatkan nilai merah dari pixel di koordinat (x, y)
            red += nilai # Menambahkan nilai ke nilai merah tersebut
            if red > 255: # Memastikan nilai merah tidak melebihi 255 (nilai maksimum)
                red = 255
            if red < 0: # Memastikan nilai merah tidak kurang dari 0 (nilai minimum)
                red = 0
            green = img[y][x][1] # Mendapatkan nilai hijau dari pixel di koordinat (x, y)
            green += nilai # Menambahkan nilai ke nilai hijau tersebut
            if green > 255: # Memastikan nilai hijau tidak melebihi 255 (nilai maksimum)
                green = 255
            if green < 0: # Memastikan nilai hijau tidak kurang dari 0 (nilai minimum)
                green = 0
            blue = img[y][x][2] # Mendapatkan nilai biru dari pixel di koordinat (x, y)
            blue += nilai # Menambahkan nilai ke nilai biru tersebut
            if blue > 255: # Memastikan nilai biru tidak melebihi 255 (nilai maksimum)
                blue = 255
            if blue < 0: # Memastikan nilai biru tidak kurang dari 0 (nilai minimum)
                blue = 0
            img_rgbbrightness[y][x] = (red, green, blue) # Memberikan nilai RGB yang sudah diubah ke pixel di koordinat (x, y)
```

Menampilkan beberapa hasil dengan nilai brightness -100 dan 100

```python
rgbbrighter(-100) # Menerapkan fungsi rgbbrighter dengan memberikan nilai brightness sebesar -100
plt.imshow(img_rgbbrightness) # Menampilkan gambar dengan perubahan brightness pada img_rgbbrightness
plt.title("Brightness -100") # Mengatur judul dari plot yang akan ditampilkan
plt.show() # Menampilkan plot

rgbbrighter(100) # Menerapkan fungsi rgbbrighter dengan memberikan nilai brightness sebesar 100
plt.imshow(img_rgbbrightness) # Menampilkan gambar dengan perubahan brightness pada img_rgbbrightness
plt.title("Brightness 100") # Mengatur judul dari plot yang akan ditampilkan
plt.show() # Menampilkan plot
```

# Contrass

1. Membuat variabel img_contrass untuk menampung hasil

```python
img_contrass = np.zeros(img.shape, dtype=np.uint8)
```

2. Melakukan penambahan contrass dengan nilai yg menjadi parameter

```python
def contrass(nilai): # Fungsi untuk mengubah kontras gambar grayscale
    for y in range(0, img_height): # Melakukan perulangan sebanyak tinggi gambar
        for x in range(0, img_width): # Melakukan perulangan sebanyak lebar gambar
            red = img[y][x][0] # Menyimpan nilai red dari gambar pada koordinat (x, y)
            green = img[y][x][1] # Menyimpan nilai green dari gambar pada koordinat (x, y)
            blue = img[y][x][2] # Menyimpan nilai blue dari gambar pada koordinat (x, y)
            gray = (int(red) + int(green) + int(blue)) / 3 # Menghitung nilai grayscale dari gambar pada koordinat (x, y)
            gray *= nilai # Mengalikan nilai grayscale dengan nilai kontras yang diinputkan
            if gray > 255: # Jika nilai grayscale melebihi 255, maka ubah menjadi 255
                gray = 255
            img_contrass[y][x] = (gray, gray, gray) # Menyimpan nilai grayscale yang telah diubah ke dalam array gambar
```

Menampilkan beberapa hasil dengan nilai contrass 50 dan 100

```python
contrass(2) # Set nilai kontras sebesar 2
plt.imshow(img_contrass) # Menampilkan gambar hasil kontras
plt.title("Contrass 2") # Memberikan judul pada gambar
plt.show() # Menampilkan gambar

contrass(3) # Set nilai kontras sebesar 3
plt.imshow(img_contrass) # Menampilkan gambar hasil kontras
plt.title("Contrass 3") # Memberikan judul pada gambar
plt.show() # Menampilkan gambar
```

# Contrass Autolevel

1. Membuat variabel img_contrass untuk menampung hasil

```python
img_autocontrass = np.zeros(img.shape, dtype=np.uint8)
```

2. Melakukan penambahan contrass secara otomatis

```python
def autocontrass(): # Mendefenisikan Untuk menyesuaikan kontras otomatis pada citra
    xmax = 300   # nilai awal xmax yang besar agar saat looping pasti lebih kecil dari gray yang dihasilkan
    xmin = 0     # nilai awal xmin yang kecil agar saat looping pasti lebih besar dari gray yang dihasilkan
    d = 0        # inisialisasi nilai d dengan 0

    # Mendapatkan nilai d, dimana nilai d ini akan berpengaruh pada hitungan
    # untuk mendapatkan tingkat kontras
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]      # ambil nilai merah di posisi (y,x)
            green = img[y][x][1]    # ambil nilai hijau di posisi (y,x)
            blue = img[y][x][2]     # ambil nilai biru di posisi (y,x)
            gray = (int(red) + int(green) + int(blue)) / 3   # konversi warna RGB ke grayscale
            if gray < xmax:       # cari nilai minimal dari grayscale
                xmax = gray
            if gray > xmin:       # cari nilai maksimal dari grayscale
                xmin = gray
    d = xmin - xmax    # hitung nilai d

    # Konversi nilai gray untuk setiap piksel berdasarkan nilai d yang telah didapat
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]      # ambil nilai merah di posisi (y,x)
            green = img[y][x][1]    # ambil nilai hijau di posisi (y,x)
            blue = img[y][x][2]     # ambil nilai biru di posisi (y,x)
            gray = (int(red) + int(green) + int(blue)) / 3   # konversi warna RGB ke grayscale
            gray = int(float(255 / d) * (gray - xmax))       # hitung nilai gray baru berdasarkan d, xmax, dan nilai grayscale
            img_autocontrass[y][x] = (gray, gray, gray)      # simpan nilai grayscale baru pada posisi (y,x) untuk setiap channel warna

    # Menampilkan gambar hasil kontras
    plt.imshow(img_autocontrass)    # menampilkan gambar hasil kontras
    plt.title("Contrass Autolevel") # memberikan judul pada gambar
    plt.show()                      # menampilkan gambar
```

3. Menampilkan hasil autolevel contrass

```python
autocontrass() # Untuk menyesuaikan kontras otomatis pada citra
plt.imshow(img_autocontrass) # Menampilkan gambar hasil kontras
plt.title("Contrass Autolevel") # Memberikan judul pada gambar
plt.show() # Menampilkan gambar
```
